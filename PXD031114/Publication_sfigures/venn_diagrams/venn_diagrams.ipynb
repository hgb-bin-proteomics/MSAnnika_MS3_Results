```python
def get_crosslink(dataset_row):
    seq_a = dataset_row["Sequence A"]
    seq_b = dataset_row["Sequence B"]
    seq_a_clean = seq_a.replace("[", "").replace("]", "")
    seq_b_clean = seq_b.replace("[", "").replace("]", "")
    
    return "-".join(sorted([seq_a_clean, seq_b_clean]))
```

```python
def get_tp_fp(row):
    if "escherichia coli" in row["Protein Descriptions A"].lower() or "escherichia coli" in row["Protein Descriptions B"].lower():
        return "fp"
    else:
        return "tp"

def get_cutoff_for_fdr(input_df, fdr = 0.01):
    input_df["TP/FP"] = input_df.apply(lambda row: get_tp_fp(row), axis = 1)
    scores = input_df["Max. XlinkX Score"].to_list()
    f_score = 0
    for score in sorted(scores):
        df = input_df[input_df["Max. XlinkX Score"] > score]
        if df[df["TP/FP"] == "fp"].shape[0] / df.shape[0] < fdr:
            f_score = score
            break
    val_1 = input_df.shape[0]
    val_2 = input_df[input_df["TP/FP"] == "fp"].shape[0]
    val_3 = input_df[input_df["Max. XlinkX Score"] > f_score].shape[0]
    print(f"Nr. of crosslinks: {val_1}")
    print(f"Nr. of false positives: {val_2}")
    print(f"Corrected score cutoff: {f_score}")
    print(f"Nr. of crosslinks at 1% 'true' FDR: {val_3}")
    return input_df[input_df["Max. XlinkX Score"] > f_score]
```

```python
sampleA_rep1_xlinkx_xls = set(get_cutoff_for_fdr(sampleA_rep1_xlinkx).apply(lambda row: get_crosslink(row), axis = 1))
sampleA_rep2_xlinkx_xls = set(get_cutoff_for_fdr(sampleA_rep2_xlinkx).apply(lambda row: get_crosslink(row), axis = 1))
sampleB_rep1_xlinkx_xls = set(get_cutoff_for_fdr(sampleB_rep1_xlinkx).apply(lambda row: get_crosslink(row), axis = 1))
sampleB_rep2_xlinkx_xls = set(get_cutoff_for_fdr(sampleB_rep2_xlinkx).apply(lambda row: get_crosslink(row), axis = 1))
sampleC_rep1_xlinkx_xls = set(get_cutoff_for_fdr(sampleC_rep1_xlinkx).apply(lambda row: get_crosslink(row), axis = 1))
sampleC_rep2_xlinkx_xls = set(get_cutoff_for_fdr(sampleC_rep2_xlinkx).apply(lambda row: get_crosslink(row), axis = 1))
```
