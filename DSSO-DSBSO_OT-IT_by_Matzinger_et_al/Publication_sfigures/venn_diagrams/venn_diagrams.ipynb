```python
def get_crosslink(dataset_row):
    seq_a = dataset_row["Sequence A"]
    seq_b = dataset_row["Sequence B"]
    
    return "-".join(sorted([seq_a, seq_b]))
```

```python
dsbso_ototit_maxlinker_xls = set(dsbso_ototit_maxlinker.loc[dsbso_ototit_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsbso_ototit_ms2annika_xls = set(dsbso_ototit_ms2annika.loc[dsbso_ototit_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsbso_ototit_ms3annika_xls = set(dsbso_ototit_ms3annika.loc[dsbso_ototit_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsbso_ototit_xlinkx_xls = set(dsbso_ototit_xlinkx.loc[dsbso_ototit_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))

dsbso_ototot_maxlinker_xls = set(dsbso_ototot_maxlinker.loc[dsbso_ototot_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsbso_ototot_ms2annika_xls = set(dsbso_ototot_ms2annika.loc[dsbso_ototot_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsbso_ototot_ms3annika_xls = set(dsbso_ototot_ms3annika.loc[dsbso_ototot_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsbso_ototot_xlinkx_xls = set(dsbso_ototot_xlinkx.loc[dsbso_ototot_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))

dsso_ototit_maxlinker_xls = set(dsso_ototit_maxlinker.loc[dsso_ototit_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsso_ototit_ms2annika_xls = set(dsso_ototit_ms2annika.loc[dsso_ototit_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsso_ototit_ms3annika_xls = set(dsso_ototit_ms3annika.loc[dsso_ototit_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsso_ototit_xlinkx_xls = set(dsso_ototit_xlinkx.loc[dsso_ototit_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))

dsso_ototot_maxlinker_xls = set(dsso_ototot_maxlinker.loc[dsso_ototot_maxlinker["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsso_ototot_ms2annika_xls = set(dsso_ototot_ms2annika.loc[dsso_ototot_ms2annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsso_ototot_ms3annika_xls = set(dsso_ototot_ms3annika.loc[dsso_ototot_ms3annika["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
dsso_ototot_xlinkx_xls = set(dsso_ototot_xlinkx.loc[dsso_ototot_xlinkx["Within same group"]].apply(lambda row: get_crosslink(row), axis = 1))
```
